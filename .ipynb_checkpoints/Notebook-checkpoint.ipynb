## Step 3: Load the Twitter US Airline Sentiment Data

Let's load our dataset and explore its structure. We'll use caching to improve performance.

````python
# Define the data path
DATA_URL = "Tweets.csv"

# Load the data with caching (for Streamlit)
# @st.cache_data decorator ensures data is loaded only once
def load_data():
    data = pd.read_csv(DATA_URL)
    data['tweet_created'] = pd.to_datetime(data['tweet_created'])
    return data

# Load the dataset
data = load_data()

# Explore the data
print(f"Dataset shape: {data.shape}")
print(f"\nColumn names: {data.columns.tolist()}")
print(f"\nFirst few rows:")
data.head()
````

````python
# Check sentiment distribution
print("Sentiment Distribution:")
print(data['airline_sentiment'].value_counts())

# Check available airlines
print("\nAvailable Airlines:")
print(data['airline'].unique())
````

## Step 5: Plot Interactive Bar Plots and Pie Charts

Visualize the number of tweets by sentiment using both bar charts and pie charts.

````python
# Prepare sentiment count data
sentiment_count = data['airline_sentiment'].value_counts()
sentiment_count = pd.DataFrame({
    'Sentiment': sentiment_count.index, 
    'Tweets': sentiment_count.values
})

print("Sentiment Counts:")
print(sentiment_count)
````

````python
# Create a bar chart
fig_bar = px.bar(
    sentiment_count, 
    x='Sentiment', 
    y='Tweets', 
    color='Tweets', 
    height=500,
    title="Number of Tweets by Sentiment (Bar Chart)"
)
fig_bar.show()
````

````python
# Create a pie chart
fig_pie = px.pie(
    sentiment_count, 
    values='Tweets', 
    names='Sentiment',
    title="Number of Tweets by Sentiment (Pie Chart)"
)
fig_pie.show()
````

## Step 6: Plotting Location Data on an Interactive Map

Display tweets on a map based on the time of day they were posted.

````python
# In Streamlit, users can select the hour
# st.sidebar.subheader("When and where are user tweeting from?")
# hour = st.sidebar.number_input("Hour of day", min_value=1, max_value=24)

# Let's filter data for a specific hour (e.g., 15:00)
hour = 15
modified_data = data[data['tweet_created'].dt.hour == hour]

print(f"Number of tweets between {hour}:00 and {(hour + 1) % 24}:00: {len(modified_data)}")
print(f"\nSample of location data:")
modified_data[['tweet_created', 'tweet_lat', 'tweet_lon', 'text']].head()
````

````python
# Visualize on a scatter map (Streamlit uses st.map() for simpler mapping)
# Here we'll use plotly for a richer visualization

# Filter out rows with missing coordinates
map_data = modified_data.dropna(subset=['tweet_lat', 'tweet_lon'])

if len(map_data) > 0:
    fig_map = px.scatter_geo(
        map_data,
        lat='tweet_lat',
        lon='tweet_lon',
        hover_name='airline',
        hover_data=['airline_sentiment', 'text'],
        color='airline_sentiment',
        title=f"Tweets between {hour}:00 and {(hour + 1) % 24}:00",
        scope='usa'
    )
    fig_map.show()
else:
    print("No location data available for this hour.")
````

## Step 7: Plot Number of Tweets by Sentiment for Each Airline

Break down sentiment analysis by individual airlines.

````python
# In Streamlit, users can select multiple airlines
# choice = st.sidebar.multiselect('Pick airlines', 
#     ('US Airways','American', 'United', 'Southwest', 'Delta', 'Virgin America'))

# Let's analyze a few airlines
choice = ['United', 'American', 'Delta']
choice_data = data[data.airline.isin(choice)]

print(f"Analyzing {len(choice)} airlines: {choice}")
print(f"Total tweets: {len(choice_data)}")
````

## Step 8: Word Cloud for Positive, Neutral, and Negative Tweets

Generate word clouds to visualize the most common words in tweets of different sentiments.

````python
# Function to generate and display word cloud
def generate_wordcloud(sentiment_type, data):
    """
    Generate a word cloud for a specific sentiment.
    
    Parameters:
    - sentiment_type: 'positive', 'negative', or 'neutral'
    - data: DataFrame containing tweet data
    """
    # Filter data based on sentiment
    df = data[data['airline_sentiment'] == sentiment_type]
    
    # Join all the tweet text into a single string
    words = ' '.join(df['text'])
    
    # Remove unwanted words (URLs, mentions, and RTs)
    processed_words = ' '.join([
        word for word in words.split() 
        if 'http' not in word 
        and not word.startswith('@') 
        and word != 'RT'
    ])
    
    # Generate the word cloud
    wordcloud = WordCloud(
        stopwords=STOPWORDS, 
        background_color='white', 
        height=640, 
        width=800
    ).generate(processed_words)
    
    # Create a figure and plot the word cloud
    fig, ax = plt.subplots(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(f'Word Cloud for {sentiment_type.capitalize()} Sentiment', fontsize=16)
    plt.xticks([])
    plt.yticks([])
    plt.tight_layout()
    plt.show()
    
    return wordcloud
````

````python
# Generate word cloud for positive tweets
print("Generating word cloud for POSITIVE tweets...")
wc_positive = generate_wordcloud('positive', data)
````

````python
# Generate word cloud for negative tweets
print("Generating word cloud for NEGATIVE tweets...")
wc_negative = generate_wordcloud('negative', data)
````

````python
# Generate word cloud for neutral tweets
print("Generating word cloud for NEUTRAL tweets...")
wc_neutral = generate_wordcloud('neutral', data)
````

## Step 9: Running the Complete Streamlit App

Now that we've explored all the components, let's understand how to run the complete dashboard:

### To run the Streamlit app:

1. Make sure all required libraries are installed:
   ```bash
   pip install -r requirements.txt
   ```

2. Run the Streamlit app from the command line:
   ```bash
   streamlit run app.py
   ```

3. Your default browser will open automatically with the interactive dashboard!

### Key Streamlit Concepts:

- **`st.title()`** - Creates a title for your app
- **`st.sidebar`** - Creates interactive sidebar elements
- **`st.markdown()`** - Renders markdown text
- **`@st.cache_data`** - Caches data to improve performance
- **`st.plotly_chart()`** - Displays interactive Plotly visualizations
- **`st.map()`** - Displays location data on a map
- **`st.pyplot()`** - Displays matplotlib figures
- **Interactive widgets**: `st.radio()`, `st.selectbox()`, `st.multiselect()`, `st.number_input()`, `st.checkbox()`
